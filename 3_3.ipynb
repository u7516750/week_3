```python
plt.suptitle('Central density vs central temperature for model 1 star')
plt.scatter(g_1_1['log_center_Rho'], g_1_1['log_center_T'], c=g_1_1['star_age'])
plt.xlabel('central density p_c(g*cm^-3)')
plt.ylabel('log T_c(K)')
plt.colorbar(label='star age(yr)')
```
