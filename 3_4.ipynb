## Temporal evolution of the surface luminosity, and the nuclear luminosities from pp, CNO, and triple-Î±

Importing the required librabries

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="ticks", palette="muted", rc={"xtick.bottom" : True, "ytick.left" : True})
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

Importing the data

```python
g= pd.read_excel("MESA.xlsx", sheet_name='3')
g.head()
```

```python
g_1= pd.read_excel("MESA.xlsx", sheet_name='5')
g_2 = pd.read_excel("MESA.xlsx", sheet_name='7')
```

```python
g.dtypes
```

```python
plt.suptitle('Central density vs central temperature for model 1 star')
plt.scatter(g['log_L'], g['pp'], c= g['star_mass'])
plt.scatter(g['log_L'], g['cno'], c= g['star_mass'])
plt.scatter(g['log_L'], g['tri_alfa'], c= g['star_mass'])
plt.xlabel('log (L_surface)')
plt.ylabel('log nuclear luminosities')
plt.colorbar(label='star age(yr)')
```

Surface Luminosity vs nuclear luminosities for 1 M_sun, and z= 0.01

```python
plt.title('Surface Luminosity vs nuclear luminosities for 1 M_sun, and z= 0.01')
plt.scatter(g['log_L'], g['pp'], label='L_pp')
plt.scatter(g['log_L'], g['cno'], label='L_cno')
plt.scatter(g['log_L'], g['tri_alfa'], label='L_alpha')
plt.ylim(-40)
#plt.xlim(,3)
plt.xlabel('log (L_surface)')
plt.ylabel('log nuclear luminosities')
plt.legend()
plt.show()
```

Checking if the nuclear luminosities addes up the surface luminosities





```python
g['L']= 10**(g['pp'])+ 10**(g['cno']) +10**(g['tri_alfa'])
```

```python
g['L_3']= 10**(g['log_L'])
```

```python
g['g']= g['L']-g['L_3']
```

```python
g.plot(y='g', use_index=True, label="difference of L_tot and L_sur")
plt.legend()
plt.show()
```
