# **Analytical solutrion for Lane-Emden Equation**

```python
from IPython.display import display, Math, Latex
import numpy as np
np.seterr(divide='ignore', invalid='ignore')

%matplotlib inline
from scipy.integrate import odeint
import matplotlib
import matplotlib.pyplot as plt
```

```python
print('Lane-Emden Equation')
display(Latex(r'$'\
              r'\dfrac{\text{d}}{\text{d}\xi}'\
              r'\left( \xi^2 \dfrac{\text{d}\theta}{\text{d}\xi} \right)'\
              r'='\
              r'-\xi^2\theta^n'\
              r'$'))
```

```python
  def G(n):
    """
    Solving Lane-Emden Equation
    input: n, int
    output: zhi_store, teta_store in array
    """
    # initialising the variables
    zhi_i = 0.00001
    d_zhi = 0.001
    dteta = 0
    teta = 1
    teta_store = [] # creating empty tetha list to store values
    zhi_store = [] # creating empty zhi list to store values
    zhi_n = zhi_i # creating a dummy variable 
    # looping through while loop
    while(teta >= 0) and (zhi_n < 20):
        zhi_n = d_zhi+ zhi_n 
        dteta_n = dteta - (((2/zhi_n)*dteta)+teta**n)*d_zhi
        teta_n = teta + dteta_n*d_zhi
        dteta = dteta_n # updating the dteta
        teta  = teta_n # updating the teta
        teta_store.append(teta) # appending teta values to list
        zhi_store.append(zhi_n) # appending the zhi values to list
    zhi_store = np.array(zhi_store)
    teta_store = np.array(teta_store)    
    return (zhi_store, teta_store)

```

assign variables to function result

```python
x_1, y_1 = G(0)
x_2, y_2 = G(1)
x_3, y_3= G(2)
x_4, y_4 = G(3)
x_5, y_5 = G(4)
x_6, y_6 = G(5)
```

Let's plot the numerical solution for n= 1,2

```python
plt.plot(x_1, y_1, label = "n=0") 
plt.plot(x_2, y_2, label = "n=1")
# setting limit
plt.ylim(0)
plt.xlim(0, 20)
# labelling axis
plt.xlabel('zhi(Dimensionless)')
plt.ylabel('teta(Dimensionless)')
plt.title('Numerical Solution of Lame Emden Equation for n=0,1')
plt.legend()
plt.show()

```
